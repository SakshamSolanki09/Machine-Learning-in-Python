```python
import pandas as pd
train_data = pd.read_csv("/content/sample_data/california_housing_train.csv")
```

```python
# Normalize Data
norm_data = (train_data-train_data.mean())/train_data.std()
```

```python
norm_data
```

```python
# Seperate X and Y features
X_train = norm_data.iloc[:,1:]
Y_train = norm_data.iloc[:,0]
```

```python
import tensorflow as tf
from keras.layers import Input,Dense,BatchNormalization
from keras.models import Sequential
```

```python
# A function to Init Neural Net
def init_network():
  net = Sequential()
  net.add(Input(shape=(8,)))
  net.add(Dense(units=8,activation="relu"))
  net.add(Dense(units=8,activation = "relu"))
  net.add(Dense(units=1,activation = "relu"))
  net.compile(loss='mse')
  return net
```

```python
net=init_network()
net.summary()
```

```python
#Train data with 500 batch size and 100 complete cycles over complete data(epochs)
net.fit(x = X_train,y = Y_train,batch_size=500,epochs=100)
```

```python
test_data = pd.read_csv("/content/sample_data/mnist_test.csv")
```

```python
#Predict new data with the trained network
net.predict(x=X_train)
```
