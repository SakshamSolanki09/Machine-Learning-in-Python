```python
import pandas as pd
import numpy as np
```

```python
# import training na testing data
train_data = pd.read_csv("/content/Devnagri_TrainingDataset.csv",index_col=0)
test_data = pd.read_csv("/content/Devnagri_TestDataset.csv",index_col=0)
```

```python
train_data
```

```python
# Seperate X and Y for both training and testing data
train_X = train_data.iloc[:,0:-1]
test_X = test_data.iloc[:,0:-1]

train_Y = train_data.iloc[:,-1]
test_Y = test_data.iloc[:,-1]
```

```python
train_Y
```

```python
# Encode the labels with numeric value instead of literal labels
encoded_train_Y = train_Y.replace(sorted(train_Y.unique()),list(range(len(train_Y.unique()))))
encoded_test_Y = test_Y.replace(sorted(train_Y.unique()),list(range(len(train_Y.unique()))))
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,Input,Activation
```

```python
# One Hot Encode the numeric labels so we can use 46 different output neurons for prediction
ohe_train_Y = keras.utils.to_categorical(encoded_train_Y)
ohe_test_Y = keras.utils.to_categorical(encoded_test_Y)
```

```python
#function to init the neural net object
def createNN():
  nn = Sequential()
  nn.add(Input(shape=(1024,)))
  nn.add(Dense(units=1024,activation="relu"))
  nn.add(Dense(units = 46,activation ="softmax"))
  nn.compile(loss = keras.losses.CategoricalCrossentropy(),metrics = ["accuracy",keras.metrics.Precision(),keras.metrics.Recall()])
  return nn
```

```python
# init neural net Object
nn = createNN()
```

```python
# Train the Network with X and one hot encoded Y, use multiprocessing and 10 workers for swift execution use testing data as validation data
nn.fit(train_X,ohe_train_Y,epochs=10,use_multiprocessing = True, batch_size= 680, validation_data=(test_X,ohe_test_Y),workers=10)
```
