```python
import pandas as pd
import numpy as np
import keras
```

```python
data = pd.read_csv("/content/drive/MyDrive/Copy of data.csv") # Read data from csv
```

```python
data
```

```python
data.drop(columns=data.columns[0],inplace = True) #Drop ID column
```

```python
# Split the data into test and training portions (1:1)
test_data = data.iloc[:data.shape[0]//2,:]
train_data = data.iloc[data.shape[0]//2:,:]

#Shuffle training data to reduce chances of overfitting 
train_data = train_data.sample(frac=1)    
```

```python
#Drop Target feature and Standardize Input features.
X = train_data.drop(columns = data.columns[0],inplace=False)
X/X.corr()
X-X.mean()
```

```python
Y = train_data.iloc[:,0]
# replacing literal labels with numeric labels for Target Feature
Y.replace(sorted(Y.unique()),list(range(len(Y.unique()))),inplace=True)
```

```python
X
```

```python
from keras import Sequential
from keras.layers import Input, Dense
```

```python
#Function to Initialize a Neural Network Model and its parameters
def createNN():
  net = Sequential()
  net.add(Input(shape = (30,)))
  net.add(Dense(units = 30,activation = "relu"))
  net.add(Dense(units = 1, activation = "sigmoid"))
  net.compile(loss=keras.losses.BinaryCrossentropy(),metrics=["Accuracy",keras.metrics.Precision(),keras.metrics.Recall()])
  return net
```

```python
# Init The Network object
nn = createNN()
```

```python
# Train the network
nn.fit(x=X,y=Y,epochs=200,use_multiprocessing = True)
```

```python
#Split Test data into X and Y
test_X = test_data.iloc[:,1:]
test_Y = test_data.iloc[:,0]

# Replace literal labels in test y with numeric labels
test_Y.replace(sorted(test_Y.unique()),list(range(len(test_Y.unique()))),inplace=True)

# Normalize the input features for test X
test_X/test_X.corr()
test_X-test_X.mean()
```

```python
# Calculate predictions with test features
pred = np.round(nn.predict(test_X))
```

```python
comp = pd.DataFrame({"pred":np.reshape(pred,(pred.shape[1],(pred.shape[0])))[0],"true":test_Y})
#Check Error squared loss
(comp.pred-comp.true).pow(2).sum()
```
